```python
# finding rolling averages for the grouped data for bayern
```

```python
def rolling_averages(group_bayern_munich, cols, new_cols):
    group_bayern_munich = group_bayern_munich.sort_values("Date")
    rolling_stats = group_bayern_munich[cols].rolling(3, closed='left').mean()
    group_bayern_munich[new_cols] = rolling_stats
    group_bayern_munich = group_bayern_munich.dropna(subset=new_cols)
    return group_bayern_munich
```

```python
def rolling_averages(group_borussia_dortmund, cols, new_cols):
    group_borussia_dortmund = group_borussia_dortmund.sort_values("Date")
    rolling_stats = group_borussia_dortmund[cols].rolling(3, closed='left').mean()
    group_borussia_dortmund[new_cols] = rolling_stats
    group_borussia_dortmund = group_borussia_dortmund.dropna(subset=new_cols)
    return group_borussia_dortmund
```

```python
def rolling_averages(group_borussia_dortmund_final, cols, new_cols):
    group_borussia_dortmund_final = group_borussia_dortmund_final.sort_values("Date")
    rolling_stats = group_borussia_dortmund_final[cols].rolling(3, closed='left').mean()
    group_borussia_dortmund_final[new_cols] = rolling_stats
    group_borussia_dortmund_final = group_borussia_dortmund_final.dropna(subset=new_cols)
    return group_borussia_dortmund_final
```
